# Imports

```
import csv
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

# Load Sentiment CSV

```
df_senti = pd.read_csv('/content/drive/MyDrive/Uni/IUI/Data/products_sentiment.csv')
df_senti.head()
```

# Load Metadata Sentiment

```
df_metadata = pd.read_csv('/content/drive/MyDrive/Uni/IUI/Data/final_dataset_metadata.csv')
df_metadata.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'], inplace = True)
df_metadata.head()
```

### Merge CSVs

```
df_final = df_metadata.merge(df_senti, left_on='asin', right_on='asin')
df_final.drop(columns=['Unnamed: 0', 'category_x'], inplace = True)
df_final.rename(columns={'category_y': 'category'}, inplace= True)
df_final.head()
```

### Minor Cleanup

```
categories = ['Office_Products', 'Appliances', 'Digital_Music', 'Pet_Supplies', 'Gift_Cards', 'Magazine_Subscriptions', 'Software']
category = 0
indexNames = df_final[df_final['category'] == categories[category]].index
df_final.drop(indexNames , inplace=True)
df_final.head()
```

# Export to final CSV

```
df_final.to_csv('/content/drive/MyDrive/Uni/IUI/Data/final_dataset_metadata_sentiment.csv')
```
